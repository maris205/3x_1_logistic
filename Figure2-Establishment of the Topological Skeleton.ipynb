```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. Parameter Calculation (Analytical)
# ==========================================
# Solving for mu in the super-stable period-3 window: x^3 - 2x^2 + x - 1 = 0
roots = np.roots([1, -2, 1, -1])
mu = np.real(roots[np.isreal(roots)])[0]

# Solving for fixed point x*: 1 - mu*x^2 = x => mu*x^2 + x - 1 = 0
x_star = (-1 + np.sqrt(1 + 4*mu)) / (2*mu)

# ==========================================
# 2. Plotting Data Preparation
# ==========================================
x_vals = np.linspace(-1.1, 1.1, 1000)
y_vals = 1 - mu * x_vals**2
diagonal = x_vals

# ==========================================
# 3. High-Quality Academic Plotting
# ==========================================
plt.figure(figsize=(10, 8))

# Plot the map f(x) and the diagonal y=x
plt.plot(x_vals, y_vals, color='black', linewidth=2.5, label=rf'$f(x) = 1 - \mu x^2$ ($\mu \approx {mu:.4f}$)')
plt.plot(x_vals, diagonal, color='gray', linestyle='--', linewidth=1.5, label=r'Diagonal $y=x$')

# --- Core: Customized Markov Partition Shading ---
# State 0 (Contracting / Even) - Blue shaded area
plt.fill_between(x_vals, -1.2, 1.2, where=(x_vals < x_star), color='#4169E1', alpha=0.12, 
                 label='State 0: Contracting (Even)')
# State 1 (Expanding / Odd) - Red shaded area
plt.fill_between(x_vals, -1.2, 1.2, where=(x_vals >= x_star), color='#DC143C', alpha=0.12, 
                 label='State 1: Expanding (Odd)')

# Vertical threshold line at x*
plt.axvline(x=x_star, color='crimson', linestyle='--', linewidth=2.5, zorder=5)

# Highlight the fixed point (x*, x*)
plt.scatter([x_star], [x_star], color='crimson', s=120, edgecolor='black', zorder=6)

# --- English Annotations ---
# Annotate fixed point x*
plt.annotate(rf'Unstable Fixed Point $x^* \approx {x_star:.4f}$', 
             xy=(x_star, x_star), xytext=(x_star + 0.1, x_star - 0.3),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
             fontsize=12, fontweight='bold')

# Region Labels
plt.text(-0.5, -0.9, 'STATE 0\n(Contracting / Even)', fontsize=14, ha='center', 
         fontweight='bold', color='#2E4053', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
plt.text(0.82, -0.9, 'STATE 1\n(Expanding / Odd)', fontsize=14, ha='center', 
         fontweight='bold', color='#7B241C', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

# Figure Aesthetics
plt.title(r'Customized Markov Partition based on the Unstable Fixed Point $x^*$', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel(r'Current State $x_n$', fontsize=13)
plt.ylabel(r'Next State $x_{n+1}$', fontsize=13)

plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.grid(True, linestyle=':', alpha=0.4)
plt.legend(loc='upper left', fontsize=11, framealpha=0.9)

plt.tight_layout()

# Save High-Resolution Image
filename = 'figure1_custom_partition_en.png'
plt.savefig(filename, dpi=300)
print(f"🎉 Figure 1 (English version) has been generated and saved as: {filename}")
plt.show()
```
